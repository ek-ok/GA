```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df = pd.read_pickle('data/beer_advocate_df.pickle')
```

```python
df.head()
```

```python
rate_cols = df.columns[4:9]
rate_df = df.groupby(df.index)[rate_cols].mean()
rate_df.head()
```

```python
beer_df = df[df.columns[:4]].drop_duplicates()
beer_df.head()
```

```python
beer_df = beer_df.join(rate_df)
beer_df.head()
```

```python
df_copy = df
```

```python
beer_df['cnt'] = df.groupby(df.index)['beer_name'].count()
beer_df.head()
```

```python
# Distribution of review counts
beer_df.cnt.plot(kind='box')
print beer_df.cnt.describe()
```

```python
beer_df['rank'] = beer_df.review_overall.rank(ascending=False).astype('int')
beer_df.head()
```

```python
beer_df.sort('rank', ascending=True)[:10]
```

```python
beer_df[rating_cols].describe()
```

```python
beer_df[['beer_name', 'review_overall', 'cnt']].sort('cnt', ascending=False)[:10]
```
