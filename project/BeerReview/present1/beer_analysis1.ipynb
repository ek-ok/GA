#Beer Review Analysis
Already did lots of data munging/transformation and saved the dataframe as pickle object

```python
df = pd.read_pickle(r'../data/beer_advocate_df.pickle')
df.head()
```

#Explore the data 

- Each row represents a beer review
- Transform the dataframe by beer taking the mean of rating columns
- Although this is one dataframe, it actually has two kinds of data sets.
    - Beer info: beer_name	brewer_ID	beer_ABV	beer_styl
    - Review info: review_appearance	review_aroma	review_palate	review_taste	review_overall

```python
rate_cols = df.columns[4:9]
rate_df = df.groupby(df.index)[rate_cols].mean()

beer_df = df[df.columns[:4]].drop_duplicates()
beer_df = beer_df.join(rate_df)

print 'Ratings by beer'
beer_df.head()
```

```python
print 'Total number of reviews:', len(df.index)
print 'Total number of beers:', len(beer_df.index)
```

# Top beers by overall ratings

Populor beers should have some reviews. For example, if there is only one review, that beer might not be populor amoung people. So create "review count" column to beer_df and analyze it first.

### Review count analysis

```python
beer_df['cnt'] = df.groupby(df.index)['beer_name'].count()
print beer_df.cnt.describe()
beer_df.cnt.plot(kind='box')
```

This boxplot did not come out nicely, which is actually interesting. This shows that the distribution of "review count" is extremely long tai. Its mean is 24 but max is 3290. 25% of the beers only have 1 review. I think this indicates that beer lovers tend to show off, meaning that they like to comment on the beers that are not commom and show their deep beer knowledge.

### Top beers

For this popularity analysis, choose the beers that have more than the mean of the review count.

```python
# Only choose the beers that have more than the mean of review count
top_5 = beer_df[beer_df.cnt >= 24].sort('review_overall', ascending=False).head()
top_5
```

#Rating Analysis

- Users could choose any ratings.
- For example it is  possible to rate
    - appearance = 1
    - aroma = 1
    - palate = 1
    - taste = 1
    - overall = 5
- So it is good idea to look at the distribution

```python
print 'Distribution of beer ratings'
df[rate_cols].describe()
```

Result: distribution is about the same across rating fields

### Look at top 5 beers

Make the ratings extreme to see how rating fields affect overall rating

```python
top_5[rate_cols].apply(lambda x: x ** 10).plot(kind='bar').legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

### Stacked bar chart except overall rating column for top 5 beers
If the other rating fields were the the "true" breakdown of overall rating, the hight of bars should be the overall scores

```python
top_5[rate_cols[:-1]].plot(kind='bar', stacked=True).legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

### Correration of ratings
Above two plots show that some columns are more important to overall ratings. So will look at more details

```python
corr = df[rate_cols].corr()
cols = [col.split('review_')[1] for col in df[rate_cols].columns]

plt.figure(figsize=(8, 6), dpi=80)
plt.imshow(corr, cmap = 'Greens', interpolation = 'none')    # Visualize the values in the entire data frame.
plt.colorbar()
plt.xticks(range(len(corr)), cols)
plt.yticks(range(len(corr)), cols);
plt.title("Correlation of Ratings")
```

Taste has the strongest correration with overall rating and appearance has the least. This means that people think taste is the most important feature of beers.

#ABV (alcohol by volume) analysis
My hypothesis is that people who really like beers tend to say they like string beers rather than light easy to drink beers. So there might be some relatinship between ABV and overall score.

```python
beer_df['beer_ABV'].describe()
```

Mean of ABV is 6.27% which is what I expected. However did not imagine that there was a beer with 57.7% of ABV.

### Plot ABV and Rating
size of the circles indicate review counts for a particulor beer

```python
abv = beer_df[(5.0 < beer_df.beer_ABV) & (beer_df.beer_ABV < 7.2)]

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.scatter(abv.beer_ABV, abv.review_overall, s=abv.cnt, alpha=0.05, c='b')
plt.title("ABV and Rating")
```

Do not see nice linear relationship as expected
