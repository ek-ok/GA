#Beer Review Analysis

```python
df = pd.read_pickle(r'/home/ikkei/Documents/GA/project/BeerReview/data/beer_advocate_clean.pickle')

df = df.set_index('beer_ID')
df = df.replace(' ', np.NaN)
df['beer_ABV'] = df['beer_ABV'].astype(float)

# df.to_pickle('beer_advocate_clean.pickle')

df.head()
```

#Explor the data 

```python
print 'Num of beers:', len(df.index)
```

```python
rating_cols = ['review_appearance', 'review_aroma', 'review_palate', 'review_taste', 'review_overall']
```

```python
df[rating_cols].describe()
```

Distribution is about the same across rating fields

# Top beers

```python
beer_unique = df[['beer_name', 'brewer_ID', 'beer_style', 'beer_ABV']].drop_duplicates()
beer_ratings = df.groupby(df.index)[rating_cols].mean()
beer_unique_ratings = beer_unique.join(beer_ratings)
beer_unique_ratings.head()
```

With review count 

```python
beer_unique_ratings['cnt'] = df.groupby(df.index)['beer_name'].count()
beer_unique_ratings.sort('review_overall', ascending=False).head()
```

```python
beer_unique_ratings['cnt'].describe()
```

```python
# Only show the beers that had more than 24 reviews
sorted = beer_unique_ratings[beer_unique_ratings.cnt > 24].sort('review_overall', ascending=False)
subset = sorted[:5]
```

```python
subset
```

#Rating Analysis

```python
subset[rating_cols].apply(lambda x: x ** 10).plot(kind='bar').legend(loc='center left', bbox_to_anchor=(1, 0.5))
subset[rating_cols[:-1]].plot(kind='bar', stacked=True).legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
corr = beer_unique_ratings[rating_cols].corr()

cols = [col.split('review_')[1] for col in beer_unique_ratings[rating_cols].columns]

plt.figure(figsize=(8, 6), dpi=80)
plt.imshow(corr, cmap = 'Greens', interpolation = 'none')    # Visualize the values in the entire data frame.
plt.colorbar()
plt.xticks(range(len(corr)), cols)
plt.yticks(range(len(corr)), cols);
plt.title("Correlation of Ratings")
```

```python
df['beer_ABV'].describe()
```

```python
abv = beer_unique_ratings[(5.2 < beer_unique_ratings.beer_ABV) & (beer_unique_ratings.beer_ABV < 8.5)]

plt.figure(figsize=(8, 6), dpi=80)
plt.subplot(1, 1, 1)
plt.scatter(abv.beer_ABV, abv.review_overall, s=abv.cnt, alpha=0.05, c='b')
plt.title("ABV and Rating")
```
