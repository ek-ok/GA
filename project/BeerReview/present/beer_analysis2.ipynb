### Delect review language

```python
def delect_language(txt):
    """
    Identify language using stop words
    
    <Possible results>
    'danish', 'dutch', 'english', 'finnish', 'french', 'german', 'hungarian', 
    'italian', 'norwegian', 'portuguese', 'russian', 'spanish', 'swedish', 'turkish'
    
    """
    if pd.isnull(txt):
        return np.NaN, np.NaN
    
    # Tokenize
    tokens = nltk.word_tokenize(str(txt))
    words = [word.lower() for word in tokens]
    
    # Delect language
    lang_ratios = {}
    for stop_lang in stopwords.fileids():
        stopwords_set = set(stopwords.words(stop_lang))
        words_set = set(words)    
        common_elem = words_set.intersection(stopwords_set)
        lang_ratios[stop_lang] = len(common_elem)
    
    # Select a language with most count
    lang = max(lang_ratios, key=lang_ratios.get)
    return lang
    
#     Based on the identified language, remove stopwords and stemmers
#     stemmer = PorterStemmer()
#     words_clean = [stemmer.stem(w) for w in words_set if w not in set(stopwords.words(lang))]
    
#     return ' '.join(words_clean), lang
```

### Convert reviews to matrix

```python
vec = feature_extraction.text.CountVectorizer(ngram_range = (1,2))
X = vec.fit_transform(reduced.review_text)
y = reduced.review_overall
```

```python
#Convert the sparse matrix back to a numpy array to run feature selection
X_arr = X.toarray()
X_df = pd.DataFrame(X_arr, columns=vec.get_feature_names())
X_columns = list(X_df.columns)
```

```python
# Before feature selection

print X_df.shape
X_df.head(5)
```

###Feature selection

```python
from sklearn import feature_selection as f_select

sig_feats = []
pvals = []

for feature in X_columns:
    fval, pval = f_select.f_classif(X_df[feature], y)
    if pval[0] < 0.05:
        sig_feats.append(feature)
        pvals.append(pval[0])
```

```python
print 'Num of original features   :', X_df.shape
print 'Num of significant features:', len(sig_feats)
```

```python
# After feature selection

X.head()
```

```python
# Re-create X selecting the significant features
X = X_df[sig_feats]
X.reset_index(inplace=True)
y = reduced.review_overall
```

```python
# Just for efficiency purpose, reduce the data size by selecting 100 beers per style
reduced = beer_reviews.groupby('beer_style').head(100)
```

```python
def get_beer_reviews(df, beer, common_users):
    mask = (df.review_profile_name.isin(common_users)) & (df.beer_name==beer)    
    return df[mask]
```

```python
rate_cols = [u'review_appearance', u'review_aroma', u'review_palate', u'review_taste', u'review_overall']

def calculate_similarity(df, distance_algo, beer1, beer2):
    # find common reviewers
    beer_1_reviewers = df[beer_reviews.beer_name==beer1].review_profile_name.unique()
    beer_2_reviewers = df[beer_reviews.beer_name==beer2].review_profile_name.unique()
    common_reviewers = set(beer_1_reviewers).intersection(beer_2_reviewers)
        
    # get reviews
    beer_1_reviews = get_beer_reviews(df, beer1, common_reviewers)
    beer_2_reviews = get_beer_reviews(df, beer2, common_reviewers)
    
    dists = []
    for col in rate_cols:
        dists.append(distance_algo(beer_1_reviews[col], beer_2_reviews[col])[0][0])
    
    return dists
```

```python
beers = reduced.beer_name.unique()

beers = beers[:10]

dists = []
for beer1 in beers:
    print "starting", beer1
    for beer2 in beers:
        if beer1 != beer2:
            try:
                row = [beer1, beer2] + calculate_similarity(reduced, euclidean_distances, beer1, beer2)
                dists.append(row)
            except:
                pass
```

```python
dist_cols = ['appearance_dist', 'aroma_dist', 'palate_dist', 'taste_dist', 'overall_dist']
cols = ["beer1", "beer2"] + dist_cols
beer_dists = pd.DataFrame(dists, columns=cols)
beer_dists.tail()
```

```python
def calc_total_dist(dists_df, beer1, beer2, weights):
    mask = (dists_df.beer1==beer1) & (dists_df.beer2==beer2)
    ratings = dists_df[mask][['appearance_dist', 'aroma_dist', 'palate_dist', 'taste_dist', 'overall_dist']]
    dists = ratings * weights
    return dists.sum(axis=1).tolist()[0]

# Test
weights = [1, 1, 1, 1, 2]
calc_total_dist(beer_dists, ' Caldera Pale Ale', ' Caldera IPA', weights)
```

```python
my_beer = " Caldera IPA"

results = []

for b in beers:
    if my_beer!=b:
        try:
            dist = calc_total_dist(beer_dists, my_beer, b, weights)
            results.append((my_beer, b, dist))
        except IndexError:
            pass
        
sorted(results, key=lambda x: x[2])
```
