#Weekday usage by hour

```python
#Weekday
df_hr_weekday = df[df.workingday==1].groupby(['hr'])
df_hr_weekday[['registered', 'casual']].sum().plot()
```

- Resistered users: It seems like resistered users on weekdays are commuters to work.
- Casual users: I think casual users are mostly tourists.

#Weekend usage by hour

```python
df_hr_weekend = df[df.workingday==0].groupby(['hr'])
df_hr_weekend[['registered', 'casual']].sum().plot()
```
