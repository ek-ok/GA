```python
import pandas as pd
import seaborn as sns
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.options.display.mpl_style = 'default'
```

```python
columns = """1) Alcohol
2) Malic acid
3) Ash
4) Alcalinity of ash  
5) Magnesium
6) Total phenols
7) Flavanoids
8) Nonflavanoid phenols
9) Proanthocyanins
10)Color intensity
11)Hue
12)OD280/OD315 of diluted wines
13)Proline"""
```

```python
cols = ['class']
cols.extend([line.strip().split(')')[1].strip().lower() for line in columns.splitlines()])
```

```python
wine = pd.read_pickle('wine.pickle')
```

```python
wine.head()
```

```python
winegroups = wine.groupby('class')
```

```python
winegroups.describe()
```

```python
winegroups.alcohol.mean()
```

```python
winegroups['alcohol', 'ash'].mean()
```

```python
winegroups.agg({'alcohol': lambda alcohol: alcohol.nunique() / alcohol.count()})
```

```python
pivoted_wine = wine.pivot_table(values='ash',
                                index='alcohol',
                                columns='class',
                                aggfunc='count')
pivoted_wine.head()
```

```python
pivoted_wine_reset = pivoted_wine.reset_index()
pivoted_wine_reset.head()
```

```python
pd.melt(pivoted_wine_reset,
        id_vars='alcohol',
        value_vars=[1,2,3],
        value_name='count').dropna().head()
```

```python
wine['alc_rounded'] = wine.alcohol.round()
wine_newgroup = wine.groupby('alc_rounded')
```

```python
round_wine = wine.pivot_table(values='ash',
                              columns='class',
                              index='alc_rounded',
                              aggfunc='count')
round_wine = round_wine.reset_index()
round_wine
```

```python
pd.melt(round_wine,
        id_vars='alc_rounded',
        value_vars=[1,2,3],
        value_name='count').dropna()
```

```python
wine_newgroup = wine.groupby(['alc_rounded', 'class'])
wine_newgroup.ash.count()
```
