```python
import pandas.tools.rplot as rplot
    
plt.figure()
plot = rplot.RPlot(last_year, x='hr')   #df, interested in hr
plot.add(rplot.TrellisGrid(['lg','.'])) #groupby lg and . menas others
plot.add(rplot.GeomHistogram())
print plot.render(plt.gcf())
```

```python
irisdf["petal_area"] = irisdf["petal length (cm)"] * irisdf["petal width (cm)"]
```

```python
def petal_guess(x):
   if(x < 1): 
       return 0 
   elif(x<7.5): 
       return 1 
   else: 
       return 2
```

```python
irisdf['guess'] = irisdf['petal_area'].apply(lambda x: petal_guess(x))
```
