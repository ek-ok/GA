```python
url = r'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
header = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
auto_mpg = pd.read_csv(url, sep='\s+', header=0, names=header, na_values='?')
```

```python
auto_mpg.head()
```

```python
auto_mpg.describe()
```

```python
header = auto_mpg.columns[:-3]
```

```python
auto_mpg.groupby(['model_year'])[header].mean()
```

```python
auto_mpg.groupby(['model_year'])['horsepower'].describe()
```

```python
horsepower_by_year = auto_mpg.groupby(['model_year'])['horsepower'].mean()
horsepower_by_year
```

```python
years = auto_mpg.model_year.unique()

for y in years:
    auto_mpg[auto_mpg.model_year == y]['horsepower'].fillna(auto_mpg[auto_mpg.horsepower.notnull()]['horsepower'].mean())
```

```python
auto_mpg[auto_mpg.horsepower.isnull()]
```

```python
auto_mpg['manufacture'] = auto_mpg.car_name.apply(lambda x: x.split()[0])
```
