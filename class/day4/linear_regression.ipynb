```python
import statsmodels.formula.api as smf

# OLS, or ordinary least squares, takes a y (dependent variable) and X (independent variables) (formula = y ~ X)
# Below, we copy the data frame and remove the na variables, and create a single variable linear model
# to return a test statistic and p-value, to see how strong of a relationship bodyweight and brainweight have.

spat_cleaned_up = pd.DataFrame(spat) #create a copy of df
spat_cleaned_up['bodywt'].dropna(inplace=True)
spat_cleaned_up['brainwt'].dropna(inplace=True)
#spat_cleaned_up.dropna(inplace=True)
spat_cleaned_up.head()
```

```python

spat_cleaned_up['log_bodywt'] = np.log(spat_cleaned_up['bodywt'])
spat_cleaned_up['log_brainwt'] = np.log(spat_cleaned_up['brainwt'])
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(spat_cleaned_up.bodywt, spat_cleaned_up.brainwt, 'g.') #green

#This model includes y intercept, y = ax + B
model = smf.ols(formula='brainwt ~ bodywt', data=spat_cleaned_up) #plot brainwt given bodywt

results = model.fit() #Compute the resuls using the model

print 'NORMAL FIT SUMMARY'
print results.summary()

#R-squared: actual vs model
#Adj. R-squared: use this
#P value: test of randomness, the certain result is due to randomness, want low p-value, p < 0.05
#R stat is in the -1 to 1 but F-stat is not 

#Intercept = 0.0859, when bodywt=0 brainwt=0.0859


axes[1].plot(spat_cleaned_up.log_bodywt, spat_cleaned_up.log_brainwt, 'm.')

log_model = smf.ols(formula='log_brainwt ~ log_bodywt', data=spat_cleaned_up)
log_results = log_model.fit()
print 'LOG-LOG FIT SUMMARY'
print log_results.summary()

#         normal | log
# -------------------------
# Adj R   0.87   |  0.931
# y       0.085  |  -4.4454     np.exp(-4.4454) = 0.0084
# bodywt  0.0001 |  0.7652
```

```python
# Original model
print sns.lmplot(x='bodywt', y='brainwt', data=spat_cleaned_up) #share is conf interval

# New log-log model
print sns.lmplot(x='log_bodywt', y='log_brainwt', data=spat_cleaned_up)
```

```python
print sm.stats.anova_lm(log_results, typ=2)
print
print log_results.get_influence().summary_table() #prints out ver observation, more details
```
