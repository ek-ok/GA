##Normal Distribution

````python
norm = pd.DataFrame({'d': np.random.normal(size=100)})
````

````python
num_bins = np.abs(sleep_df.awake.max() - sleep_df.awake.min()) / 2
num_bins
````

````python
fig, axes = plt.subplots(nrows=2, ncols=2)

norm.d.plot(ax=axes[0, 0], kind='kde')
norm.d.hist(ax=axes[0, 1], bins=num_bins)

sleep_df.awake.plot(ax=axes[1, 0], kind='kde')
sleep_df.awake.hist(ax=axes[1, 1], bins=num_bins)
````

````python
qq = sorted(np.random.normal(sleep_df.awake.mean(), sleep_df.awake.std(), len(sleep_df))) #mean, std, N
awake = sorted(sleep_df.awake)

plt.plot(qq, awake, '.')
plt.show()
````

````python
import statsmodels.formula.api as smf

sleep_clean_df = sleep_df

sleep_clean_df.bodywt.dropna(inplace=True)
sleep_clean_df.brainwt.dropna(inplace=True)

sleep_clean_df['log_bodywt'] = np.log(sleep_clean_df['bodywt'])
sleep_clean_df['log_brainwt'] = np.log(sleep_clean_df['brainwt'])

sleep_clean_df.head()
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(sleep_clean_df.bodywt, sleep_clean_df.brainwt, 'g.')
axes[1].plot(sleep_clean_df.log_bodywt, sleep_clean_df.log_brainwt, 'b.')
````

````python
# OLS, or ordinary least squares, takes a y (dependent variable) and X (independent variables) (formula = y ~ X)
model = smf.ols(formula='brainwt ~ bodywt', data=sleep_clean_df)
results = model.fit()
print results.summary()

print
print

log_model = smf.ols(formula='log_brainwt ~ log_bodywt', data=sleep_clean_df)
log_results = log_model.fit()
print log_results.summary()
````
