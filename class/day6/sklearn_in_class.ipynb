```python
# any other imports here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from __future__ import division
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
from sklearn import feature_selection as f_select
from sklearn import linear_model as lm
from sklearn import cross_validation as cv
from sklearn import metrics
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
desc = boston.DESCR
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['MEDV'] = boston.target

df.head()
```

```python
for l in desc.splitlines():
    print l
```

```python
x_cols = list(df.columns)
y_col = 'MEDV'
x_cols.remove('MEDV')
```

```python
sig_cols = []
sig_pvals = []

for x_col in x_cols:
    # Find linear regression
    fval, pval = f_select.f_regression(df[x_col], df[y_col])
    
    fval = fval[0] # Determines how well the variance is due the expected variance.
    pval = pval[0] # 
    
    if pval < 0.05:
        sig_cols.append(x_col)
        sig_pvals.append(pval)

print 'sig_cols: ', sig_cols
print 'sig_pvals: ', sig_pvals
```

```python
x_train, x_test, y_train, y_test = cv.train_test_split(df[sig_cols],
                                                       df[y_col],
                                                       test_size=0.2,
                                                       random_state=1234)
```

```python
model = lm.LinearRegression().fit(x_train, y_train)

print pd.DataFrame({'column': sig_cols,
                    'coef': model.coef_,
                    'p-value': sig_pvals,}).set_index('column')
```

```python
# For training sets
print 'score: ', model.score(x_train, y_train)
print 'R2: ', metrics.r2_score(y_train, model.predict(x_train))
```

```python
# For testing sets
print 'score: ', model.score(x_test, y_test)
print 'R2: ', metrics.r2_score(y_test, model.predict(x_test))
```

#Calculate $R^2$

* Sum of squares due to regression
* Sum Squared Error
* Trying to minimize this
* Cost function
* $SSR = \Sigma(\hat{y_i}-\bar{y})^2$
* $\hat{y_i}$ : Target variable

```python
sum_squares_regr = np.sum((model.predict(x_train) - np.mean(y_train))**2)
print sum_squares_regr
```

* Total Sum Squares
* $SST = SSR+SSE$
* $SSR$ : Sum of squares due to regression (cost function)
* $SSE$ : Sum of Squared Residiuals (??)

```python
total_sum_squares = np.sum((y_train - np.mean(y_train))**2)
print total_sum_squares
```

$R^2 = SSR / SST$

```python
r_squared = sum_squares_regr / total_sum_squares
print r_squared
```
