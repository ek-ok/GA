```python
%pylab inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.mpl_style', 'default')

plt.rcParams['figure.figsize'] = (20, 10)
```

```python
n = 1000
z = np.random.randn(n)
walk = np.zeros(n)
```

```python
for i in np.arange(1, 1000, 1):    
    walk[i] = walk[i - 1] + z[i]
plt.plot(walk)
```
